#### Interpretation
Starting in some form of reverse order in terms of analysis performed, the most significant words surfaced by this analysis are philosophy, materialism, and religion. The significance of these words is reinforced by their separation from any sort of cluster at both the levels of the first and second principal component. As to why, my supposition is as follows. 

Materialism, as influenced by Hegelian thought and formalized by Dietzgen, was the crucial ideological tool for Engels and Marx as their works were basad in the belief that the material conditions of society - such as economic systems, modes of production, and class relations - were the primary drivers of historical development and the socio-economic conditions of any classes, explaining the presence of this word as highly significant. Secondly, religion is likely viewed as a significant word as Marx and Engels saw religion as a reflection of the material conditions of society, and argued that it could be used to maintain social order and reinforce the power of the ruling class. Moreover, Engels and Marx saw the development of religion as a response to the alienation and suffering that resulted from capitalist economic systems, explaining why it would be such focus within in their texts. Finally, the presence of philosophy as the most significant term, by a notable margin, makes sense as this entire corpus of works was either a critique of existing philosophy or the presentation of new philosophical principles.

The final point of interpretation is the HCA plot. Using jaccard distance, this plot identifies Engel's three works *The Condition of the Working-Class in England in 1844*,  *The Origin of the Family, Private Property and the State*, *Landmarks of Scientific Socialism: "Anti-Duehring"* and Marx's work *Revolution and Counter-Revolution* as most similar. Moreover, it identifies *The Communist Manifesto* as the most distinct. This visual conclusion is logical as the first four texts were direct commentaries on very specific material in certain social and economic situations, all from the same ideological stance; the Manifesto's uniqueness can be explained in its uniqueness as a text throughout time.
